```python
import numpy as np
import pandas as pd
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
salary_data = pd.read_csv('./data/2020salary.csv', encoding='EUC-KR')
salary_data.head()
```

```python
salary_data.drop(['이름'], axis=1, inplace=True)
salary_data.head()
```

```python
data = np.array(salary_data, dtype=np.float32)
print(type(data))
data
```

```python
xData = data[:, 0:6]
print(type(xData))
print(xData)
```

```python
yData = data[:, 6:] # 2차원 데이터로 뽑아낸다. => 변화 요인이 2차원이므로 결과도 2차원 형태로 뽑아내야 한다.
print(type(yData))
print(yData)
```

```python
X = tf.placeholder(dtype=tf.float32, shape=[None, 6]) 
Y = tf.placeholder(dtype=tf.float32, shape=[None, 1])
```

```python
a = tf.Variable(tf.random_uniform([6, 1]), dtype=tf.float32)
b = tf.Variable(tf.random_uniform([1]), dtype=tf.float32)
```

```python
y = tf.matmul(X, a) + b
loss = tf.reduce_mean(tf.square(y - Y))
gradient_descent = tf.train.GradientDescentOptimizer(0.00000005).minimize(loss)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for i in range(1000001):
    loss_, y_, _ = sess.run([loss, y, gradient_descent], feed_dict={X: xData, Y: yData})
    if i % 5000 == 0:
        print('Epoch: {0:6d}, loss: {1:12.3f}, price: {2:8.2f}'.format(i, loss_, y_[0][0]))

saver = tf.train.Saver()
save_path = saver.save(sess, './data/saved.cpkt')
print('학습된 모델을 저장했습니다.')
```
